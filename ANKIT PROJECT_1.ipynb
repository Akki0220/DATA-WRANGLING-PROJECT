# Check Outliers

```python
columns = ["instant","dteday","windspeed","holiday","hum","temp","casual","registered","cnt"]
final_data[columns] =final_data[columns].apply(pd.to_numeric)
```

```python
def find_outliers(df,columns):
    q1 = df[columns].quantile(.25)
    q3 = df[columns].quantile(.75)
    iqr = q3 - q1
    min_r = q1 - 1.5*iqr
    max_r = q3 + 1.5*iqr
    outliers_indices = df.index[(df[columns] < min_r) | (df[columns]> max_r)]
    return outliers_indices
```

```python
find_outliers(final_data, "cnt")
```

```python
find_outliers(final_data,"hum")
```

```python
find_outliers(final_data,"temp")
```

```python
find_outliers(final_data,"windspeed")
```
